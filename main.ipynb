```python
import spacy
import pdfplumber

pdf = pdfplumber.open("CV.pdf")
page = pdf.pages[0]

text = page.extract_text()
words = page.extract_words()

image = page.to_image()
image.draw_rects(words)
```

```python
print(text)
```

```python
nlp = spacy.load("en_core_web_trf")

doc = nlp(text)

spacy.displacy.render(doc, style = "ent")
```
